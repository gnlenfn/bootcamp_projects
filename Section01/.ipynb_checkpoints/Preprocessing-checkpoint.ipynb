### 데이터 불러오기

```python
import pandas as pd

data = pd.read_csv("vgames2.csv")
data.head()
```

```python
data.shape
```

### 결측치 처리

```python
data.info()
```

```python
data.drop("Unnamed: 0", inplace=True, axis=1)
```

```python
data.isnull().sum()
```

```python
data.dropna(axis=0, inplace=True)
data.shape
```

### 단위를 백만으로 통일

```python
data.sort_values(by='NA_Sales', ascending=False)
```

```python
sales = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']

for col in sales: 
    """
    # remove K, M and convert to float
    # seperate K, M as other column (str.extract)
    # fillna(1) for multiply
    # divide num with million
    """
    new = \
        data[col].replace(r'[KM]+$', '', regex=True).astype(float) * \
        data[col].str.extract(r'[\d\.]+([KM]+)', expand=False).\
        fillna(1).\
        replace(['K','M'], [0.001, 1/10**6]).astype(float)                     
    data[col] = new * 1_000_000
    data[col] = data[col].astype(int)
data.head()
```

### 연도별 흥행작 확인

```python
data['Year'] = data['Year'].astype(int)
data.sort_values(by='Year', ascending=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
recent.head()
```

```python
pd.pivot_table(data=recent, index=['Platform', 'Year'])
```

```python
tmp = pd.pivot_table(data=recent, index=['Platform', 'Year']).reset_index()
tmp.head()
```

```python
sns.barplot(data=tmp, x='Platform', y='NA_Sales')
```

```python
by_region = pd.melt(tmp, id_vars=["Platform", 'Year'])
```

```python
fig, ax = plt.subplots(figsize=(16,8))
fig = sns.barplot(data=by_region, x='Year', y='value', hue='variable', ci=None)
ax.set_title("Mean of Sales by Region")
```

- 북미지역 판매량이 항상 가장많음 (집중해야할 지역)

```python
fig, ax = plt.subplots(figsize=(16,8))
fig = sns.barplot(data=by_region, x='Year', y='value', hue='Platform', ci=None)
ax.set_title("Mean of Sales by Region")
```
