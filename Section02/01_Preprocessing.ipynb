# Preprocessing
Ìï®ÏàòÎ°ú ÎßåÎì§Í∏∞

```python
def calcuate_ability(string):
    if string.isnumeric():
        ability = int(string)
    else:
        nums = list(filter(None, string.split("+")))
        nums = list(map(int, nums))
        ability = sum(nums)
    return ability

def count_tags(strings):
    if type(strings) == int:
        tags = 0
    else:
        tags = len(strings.split(","))
    return tags
```

```python
def preprocessing(df):
    dirt = ['ls','st','rs','lw','lf','cf','rf','rw','lam','cam','ram','lm','lcm','cm','rcm','rm','lwb','ldm', 'cdm','rdm','rwb','lb','lcb','cb','rcb','rb']
    
    # GK abilities
    goalie = df[['goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking',
                      'goalkeeping_positioning', 'goalkeeping_reflexes']]
    goalie_ability = goalie.mean(axis=1).round().astype(int).astype(str)
      ## add player GK ability
    df['gk'] = np.where(df['player_positions'] == 'GK', goalie_ability, '0')
    position_ability = df[dirt+['gk']]
    position_ability = position_ability.fillna('0')
    for col in position_ability.columns:
        position_ability[col] = position_ability[col].apply(calcuate_ability)
      ## set position with position_ability table
      ## player_positions, team_positionÏùÑ ÎåÄÏ≤¥
    ability_of_player = position_ability.max(axis=1)
    pos = position_ability.idxmax(1)
    df['positions'] = np.where((df['team_position'] == 'SUB')|(df['team_position'] == 'RES'), pos.str.upper(), df['team_position'])
    df['posotion_ability'] = ability_of_player
    
    
    
    # league_name NaNÏù∏ ÏÑ†ÏàòÎäî ÏûêÏú†Í≥ÑÏïΩ --> drop
    idx = df['league_name'].notna()
    df = df[idx]
    
    # drop columns
    df = df.drop(['team_position', 'player_positions'], axis=1)
    df = df.drop(dirt+['gk'], axis=1)
    df = df.drop(['sofifa_id', 'player_url', 'long_name', 'dob', 'body_type', 'real_face',
                              'nation_position', 'nation_jersey_number', 'loaned_from'], axis=1)
    
    to_zero = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'gk_diving', 
              'gk_handling', 'gk_kicking', 'gk_reflexes', 'gk_speed', 'gk_positioning', 'player_traits', 
              'defending_marking', 'release_clause_eur', 'player_tags']
    drop_cols = ['joined', 'contract_valid_until', 'team_jersey_number', 'league_rank']
    df[to_zero] = df[to_zero].fillna(0)
    df = df.drop(drop_cols, axis=1)
    # fill NA --> goalkeeper ability of field players and vice versa
    df = df.fillna(0)
    
    # GK Ï†úÏô∏
    df = df[df['positions'] != 'GK']
    
    # count tags
    df['player_tags'] = df['player_tags'].apply(count_tags)
    df['player_traits'] = df['player_traits'].apply(count_tags)
    
    # add BMI
    df['bmi'] = df['weight_kg'] / ((df['height_cm'] * 0.01) ** 2)
    
    return df
```

```python
players = pd.read_csv("./data/players_20.csv")
fifa_20 = preprocessing(players)
fifa_20.to_csv("./data/fifa_20.csv", index=False)
```
